# Functions:

```python
import numpy as np


# Python code to get the Cumulative sum of a list
def Cumulative(lists):
    cu_list = []
    length = len(lists)
    cu_list = [sum(lists[0:x:1]) for x in range(0, length+1)]
    return cu_list[1:]

def bowl_outcome(prob_list):
    pl = prob_list
    outcome = ["Fare Ball","No Ball","Byes","Legbyes","Wides"]
    cpl = Cumulative(pl)
    x = []
    u = np.random.uniform(size=1)
    for i in range(0,len(cpl)):
        if u > cpl[i]:
            x.append(0)
        else:
            x.append(1)
    c = Cumulative(x)
    oc = outcome[c.index(1)]
    return(oc)
        
def bat_outcome(prob_list):
    bpl = prob_list
    outcome = [1,2,3,4,5,6,"Wicket",0]
    cbpl = Cumulative(bpl)
    y = []
    u = np.random.uniform(size=1)
    for i in range(0,len(cbpl)):
        if u > cbpl[i]:
            y.append(0)
        else:
            y.append(1)
    c = Cumulative(y)
    oc = outcome[c.index(1)]
    return(oc)



India_wickets_prob = [0.082191781,0.726027397,0.068493151,0.04109589,0.082191781]
SA_wickets_prob = [0.166666667,0.546296296,0.018518519,0.194444444,0.074074074]
def wicket_outcome(prob_wicket):
    pl = prob_wicket
    outcome = ["Bowled","Cought","Cought & bowled","LBW","Runout"]
    cpl = Cumulative(pl)
    x = []
    u = np.random.uniform(size=1)
    for i in range(0,len(cpl)):
        if u > cpl[i]:
            x.append(0)
        else:
            x.append(1)
    c = Cumulative(x)
    w = outcome[c.index(1)]
    return(w)
```
